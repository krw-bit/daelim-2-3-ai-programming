조건문

```python
temp = 8

if temp > 20:
    print("덥다.")

elif temp > 10:
    print("따뜻하다.")

elif temp > 0:
    print("춥다.")

else:
    print("매우 춥다.")
```

반복문과 조건문 함께 사용

```python
scores = [42, 60, 77, 89, 20]

for score in scores:
    if score >= 60:
        print(f'{score}점은 합격입니다.')
    
    else:
        print(f'{score}점은 불합격입니다.')

# 60점은 합격입니다.
# 42점은 불합격입니다.
```

함수

```python
def greet(name):
    print(f'{name}님, 안녕하세요.')

greet("대림대")
```

```python
def add_numbers(a, b):
    return a + b

print(add_numbers(1, 2))
```

```python
numbers = [1, 2, 3, 4, 5]
total = 0

for number in numbers:
    total += number

print("합계:", total)

# 합계: numbers를 다 더한 값
```

```python
scores = [65, 82, 91, 100, 58, 77]
count = 0

for score in scores:
    if score >= 70:
        count += 1

print(f'합격자는 {count}명 입니다.')
        
# 70점 이상은 합격
# 출력: 합격자는 x명 입니다.
```
