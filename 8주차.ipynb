```python
import pandas as pd
df = pd.read_csv('integrated_student_data.csv')
student_df = pd.DataFrame(df)

column_names = df.columns.tolist()

df['name'] = df['name'].str.strip()
df['email'] = df['email'].str.strip()
df['email'] = df['email'].str.lower()

df['domain'] = df['email'].str.split('@').str[1]
df['invalid'] = df['email'].str.split('-').str[1]

df['final_score'] = ((df['assignment_score'] + df['quiz_avg'] + df['project_score'])/3).round(1)
df['achievement_level'] = df['final_score']

print(df.head())
```
